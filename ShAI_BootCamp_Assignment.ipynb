#About Dataset
salaries dataset generally provides information about the employees of an organization in relation to their compensation. It typically includes details such as how much each employee is paid (their salary), their job titles, the departments they work in, and possibly additional information like their level of experience, education, and employment history within the organization.

# About Features

1. **Employee ID**: A unique identifier for each employee. Usually a numeric or alphanumeric code.

2. **Name**: The full name of the employee. This might be split into first and last names in some datasets.

3. **Department**: The department or division within the organization where the employee works.

4. **Job Title**: The official title or position held by the employee in the organization.

5. **Salary**: The amount of money the employee earns. This could be represented as an annual salary, hourly wage, or monthly pay.

6. **Hire Date**: The date when the employee started working at the organization. Useful for calculating tenure or experience.

7. **Gender**: The gender of the employee. This might be included in some datasets for diversity and equality analysis.

8. **Years of Experience**: The total number of years the employee has worked in their field or profession.

9. **Education Level**: The highest level of education attained by the employee, such as a high school diploma, bachelor's degree, etc.

10. **Performance Rating**: If available, this could be a measure of the employee's performance in their role, often based on annual review scores.

11. **Location**: The geographical location or office where the employee works, if the organization has multiple locations.

```python
import pandas as pd
import numpy as np

salaries_data  = pd.read_csv("SalariesX.csv")
salaries_data .head()
```

# 1.Basic Data Exploration

```python
rows, columns = salaries_data.shape

data_types = salaries_data.dtypes

missing_values = salaries_data.isnull().sum()

# Display the results
print(f"Number of Rows: {rows}")
print(f"Number of Columns: {columns}")
print("\nData Types of Each Column:")
print(data_types)
print("\nMissing Values in Each Column:")
print(missing_values)
```

# 2.Descriptive Statistics

```python
list_of_name=['BasePay' , 'OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']
for i in range(len(list_of_name)):
    
    mean_salary = salaries_data[list_of_name[i]].mean()
    median_salary = salaries_data[list_of_name[i]].median()
    mode_salary = salaries_data[list_of_name[i]].mode()[0]  
    min_salary = salaries_data[list_of_name[i]].min()
    max_salary = salaries_data[list_of_name[i]].max()
    range_salary = max_salary - min_salary
    std_dev_salary = salaries_data[list_of_name[i]].std()
    print (f"{list_of_name[i]}")
    print(f"Mean Salary: {mean_salary}")
    print(f"Median Salary: {median_salary}")
    print(f"Mode Salary: {mode_salary}")
    print(f"Minimum Salary: {min_salary}")
    print(f"Maximum Salary: {max_salary}")
    print(f"Range of Salaries: {range_salary}")
    print(f"Standard Deviation of Salary: {std_dev_salary}")
    print("------------")
```

# 3  Data Cleaning


```python
import pandas as pd
# Data Cleaning

#حذف عامود Notes  لانه فارغ وليس له تاثير ايجابي ع داتا لذلك قمنا بحذفه
salaries_data_cleaned = salaries_data.drop(columns=['Notes'])

for column in ['BasePay', 'OvertimePay', 'OtherPay']:
    median_value = salaries_data_cleaned[column].median()
    salaries_data_cleaned[column].fillna(median_value, inplace=True)
    
salaries_data_cleaned=salaries_data_cleaned.drop_duplicates()

print(salaries_data_cleaned.isnull().sum())

cleaned_file_path = 'SalariesX_Cleaned.csv'
salaries_data_cleaned.to_csv(cleaned_file_path, index=False)
```

# 4.Basic Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
# 1. Histogram for Salary Distribution
plt.figure(figsize=(10, 6))
sns.histplot(salaries_data_cleaned['BasePay'], bins=30, kde=False)
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#يوجد عدد كبير من اقسام العمل وهذا صعب تمثيله في بي جارت لذلك قمت باخذ اكثر 20 اقسام لديها عدد اكبر من العاملين
department_counts = salaries_data_cleaned['JobTitle'].value_counts().head(20)
plt.figure(figsize=(25, 25))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different JobTitle')
plt.show()
```

# 5.Grouped Analysis

```python
import pandas as pd

group_columns = ['JobTitle', 'Year']
grouped_data = salaries_data_cleaned.groupby(group_columns)
summary_stats = grouped_data['BasePay'].describe()
mean_salaries = summary_stats['mean'].sort_values(ascending=False)
# Displaying the results
print("Summary Statistics for each Group:\n")
print(summary_stats)
print("\nAverage Salaries Comparison:\n")
print(mean_salaries)
```

# 6.Simple Correlation Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt

comparison_column = 'OvertimePay'
correlation = salaries_data['BasePay'].corr(salaries_data[comparison_column])
plt.figure(figsize=(10, 6))
plt.scatter(salaries_data['BasePay'], salaries_data[comparison_column], alpha=0.5)
plt.title(f'Scatter Plot of BasePay vs. {comparison_column}')
plt.xlabel('BasePay')
plt.ylabel(comparison_column)
plt.show()
print(f"Correlation coefficient between BasePay and {comparison_column}: {correlation}")
```
